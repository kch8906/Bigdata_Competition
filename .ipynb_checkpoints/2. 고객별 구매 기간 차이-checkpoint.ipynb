### 구매 기간 확인

```python
df_02['de_dt'] = df_02['de_dt'].astype(str)
df_02['de_dt'] = pd.to_datetime(df_02['de_dt'])
df_02
```

```python
df_03['de_dt'] = df_03['de_dt'].astype(str)
df_03['de_dt'] = pd.to_datetime(df_03['de_dt'])
df_03
```

```python
df_concat_02_03 = pd.concat([df_02, df_03])
df_concat_02_03.drop('vst_dt', axis=1, inplace=True)
df_concat_02_03
```

```python
from datetime import timedelta

def period_purchasing(cust, df):
    try:
        person_id = cust
        res = df[df['cust'] == person_id]['de_dt']
        res = sorted(list(set(res)))
        if len(res) > 1:
            ans = res[-1] - res[-2]
            ans = ans.days
        else:
            ans = 0

        return ans
    
    except: IndexError
```

```python
df_01['P_period'] = df_01.apply(lambda x : period_purchasing(x['cust'], df_concat_02_03), axis=1)
```

```python
df_01_period = df_01_period[~df_01_period['P_period'].isna()]
df_01_period.drop('Unnamed: 0', axis=1, inplace=True)
```
